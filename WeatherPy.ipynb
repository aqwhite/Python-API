```python
from config import gkey
import numpy as np
import pandas as pd
import requests
import random
import matplotlib.pyplot as plt 
import seaborn
from citipy import citipy
import json
```

```python
rand_city = []
rand_country = []
rand_lat = []

while len(rand_city) < 500:
    lat_pos = random.randint(0,2)
    base_lat = random.randint(0,91)
    dec_lat = random.random()/100
    lat = base_lat + dec_lat
    if lat_pos == 1:
        lat = lat * -1
    rand_lat.append(lat)
   
    lng_pos = random.randint(0,2)
    base_lng = random.randint(0,91)
    dec_lng = random.random()/100
    lng = base_lng + dec_lng
    if lat_pos == 1:
        lng = lng * -1

    city = citipy.nearest_city(lat, lng)
    
    foundCity = False
    for j in range(len(rand_city)):
        if city.city_name == rand_city[j]:
            foundCity = True
    if foundCity==False  :   
        rand_city.append(city.city_name)
        rand_country.append(city.country_code)
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

weather_url = base_url + "appid=" + gkey  + "&units=" + units + "&q=" 
```

```python
wind = []
temp = []
humid = []
cloud = []
city_lat = []
city_name = []
count = 0
for city in rand_city:
    response = requests.get(weather_url + city).json()
    print(f"Getting informaiton for {city}")

    try:
        wind.append(response["wind"]["speed"])
        temp.append(response["main"]["temp"])
        humid.append(response["main"]["humidity"])
        cloud.append(response["clouds"]["all"])
        city_name.append(city)
        city_lat.append(response["coord"]["lat"])
        city_id = response["id"]
        print(f"Information found for {city} with city ID: {city_id}")
    except:
        count += 1
        print(f"No information was found for {city}")

print(f"There was no data gathered for {count} countries")
```

```python
print(len(temp))
print(len(wind))
print(len(humid))
print(len(cloud))
print(len(city_lat))
print(len(city_name))
```

```python
weather_dict = {
    "City" : city_name,
    "Wind" : wind, 
    "Temperature" : temp, 
    "Humidity" : humid,
    "Cloudiness" : cloud,
    "Latitude" : city_lat,
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```

```python
weather_data.to_csv("weatherdata.csv", encoding="utf-8", index=False)
```

```python
plt.scatter(weather_data["Latitude"], weather_data["Temperature"], marker="o")
plt.title("City Latitude vs Max Temperature")
plt.ylabel("Temperature (F) ")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim(-90,90)
plt.show()
plt.savefig("Lat_vs_Temperature.png")
```
